```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/suburb_score.csv')
df = df.sort_values('total_score',ascending=False)
df = df.reset_index(drop=True)


```

```python
df1 = pd.read_csv('data/hist_and_pred_median.csv')
df1['label'] = 'predicted'
df1.loc[df1['year'] <= 2024, 'label'] = 'historical'
df1 = df1[df1['suburb'] == 'Whittlesea']
historical_data = df1[df1['label'] == 'historical']
plt.plot(historical_data['year'], historical_data['median_rental_price'], marker='o', label='Historical', linestyle='-', color='blue')

# Plot the line graph for predicted data (dashed line)
predicted_data = pd.concat([df1[df1['year'] == 2024], df1[df1['label'] == 'predicted']])

plt.plot(predicted_data['year'], predicted_data['median_rental_price'], marker='o', label='Predicted', linestyle='--', color='red')

plt.title(f'Median Rental Prices in (Historical vs Predicted)')

plt.xlabel('Year')
plt.ylabel('Median Rental Price ($)')
plt.legend()

plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
